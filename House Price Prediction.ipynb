```python
columns_to_encode = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
                    'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
                    'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
                    'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
                    'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
                    'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
                    'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
                    'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature',
                    'SaleType', 'SaleCondition']

# for column in columns_to_encode:
#     label_encoder = LabelEncoder()
#     train[column] = label_encoder.fit_transform(train[column])

train = train.drop(columns=columns_to_encode)
test = test.drop(columns=columns_to_encode)
```

```python
train
```

```python
X=train.drop(columns='SalePrice')
y=train['SalePrice']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
SalePrice = model.predict(test.drop(columns='Id'))
```

```python
submission_df = pd.DataFrame({"Id": test['Id'], "SalePrice": SalePrice})
```

```python
submission_df.to_csv("LinearRegression.csv", index=False)
```
